```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data.csv", header=None)
df[len(df.columns)] = 0.0
df.loc[df.iloc[:,1] == 'M', len(df.columns) - 1 ] = 1
df.drop([0, 1], axis=1, inplace=True)
df.columns = range(len(df.columns))
benin_set = df.loc[df.iloc[:,-1] == 0, 0:29]
malin_set = df.loc[df.iloc[:,-1] == 1, 0:29]
```

# Historgram

```python
nb_col_graph = 5
nb_lignes_graph = 6
col_names = benin_set.columns
plt.figure(figsize=(15, 15))
for i in range(len(col_names)):
    plt.subplot(nb_lignes_graph, nb_col_graph, i + 1)
    plt.hist(benin_set.loc[:,col_names[i]], alpha = 0.5, lw=3, label="benin", color="b")
    plt.hist(malin_set.loc[:,col_names[i]], alpha = 0.5, lw=3, label="malin", color="y")
    plt.legend()
manager = plt.get_current_fig_manager()
manager.set_window_title("Histogram")
plt.show()
```
